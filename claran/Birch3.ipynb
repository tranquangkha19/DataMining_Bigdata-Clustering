```python
from clarans import CLARANS
import numpy as np
import csv
import time
import matplotlib.pyplot as plt
```

```python
data = []
```

```python
with open('birch3.txt', 'r') as file:
    for line in file:
        tmp = []
        x1, x2 = line.split()
        val1, val2 = int(x1), int(x2)
        tmp.append(val1)
        tmp.append(val2)
        data.append(tmp)
```

```python
x_axis = [point[0] for point in data]
y_axis = [point[1] for point in data]
```

```python
plt.scatter(x_axis, y_axis, marker='o')
```

```python
clarans_obj = CLARANS(data, 100, 1, 5)
```

```python
start = time.time()
clarans_obj.process()
end = time.time()
```

```python
clst = clarans_obj.clusters
med = clarans_obj.optimal_medoids
print(len(clst))
print(med)
print(end-start)
```

```python
for i in range(len(clst)):
    cluster = [data[idx] for idx in clst[i]]
    x_cluster = [point[0] for point in cluster]
    y_cluster = [point[1] for point in cluster]
    plt.scatter(x_cluster , y_cluster, label=i)
    plt.scatter(data[med[i]][0] , data[med[i]][1] , s = 30, color = 'k')
```
