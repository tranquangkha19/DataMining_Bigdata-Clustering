```python
from clarans import CLARANS
import numpy as np
import csv
import time
import matplotlib.pyplot as plt
```

```python
with open('datasets/10kData-10clusters.csv','r') as file:
    rows = csv.reader(file,
                      delimiter = ',',
                      quotechar = '"')
    data = [data for data in rows]

```

```python
for i in range(20):
    current = [point for point in data if int(point[2]) == i]
    x_cluster = [float(point[0]) for point in current]
    y_cluster = [float(point[1]) for point in current]
    plt.scatter(x_cluster , y_cluster, label=i)
```

```python
data2 = [[float(num) for num in item] for item in data]
```

```python
clarans_obj = CLARANS(data2, 10, 1, 5)
```

```python
start = time.time()
clarans_obj.process()
end = time.time()
```

```python
clst = clarans_obj.clusters
med = clarans_obj.optimal_medoids
print(len(clst))
print(med)
print(end-start)
```

```python
for i in range(len(clst)):
    cluster = [data2[idx] for idx in clst[i]]
    x_cluster = [point[0] for point in cluster]
    y_cluster = [point[1] for point in cluster]
    plt.scatter(x_cluster , y_cluster, label=i)
    plt.scatter(data2[med[i]][0] , data2[med[i]][1] , s = 30, color = 'k')
```
