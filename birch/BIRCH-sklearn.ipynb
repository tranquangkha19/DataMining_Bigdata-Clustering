```python
from sklearn.cluster import Birch
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# X = np.loadtxt('b1.txt',delimiter=',')
# X = np.loadtxt('b2.txt',delimiter=',')
X = np.loadtxt('b3.txt',delimiter=',')
```

```python
num_cluster = None
branching = 10000
threshold = 40000
```

```python
plt.figure(figsize=(12,12))
plt.scatter(X[:,0], X[:,1], s=3, alpha=0.1)
```

```python
brc = Birch(n_clusters=num_cluster, branching_factor=branching, threshold=threshold)

starttime = time.perf_counter()
brc.fit(X)
endtime = time.perf_counter()

print("time elapsed: ", endtime-starttime)
```
